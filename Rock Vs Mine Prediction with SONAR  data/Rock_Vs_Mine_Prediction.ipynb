```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df  = pd.read_csv('/content/Copy of sonar data.csv', header = None)
```

# Seperating the data and labels

```python
X =  df.drop(columns = 60, axis = 1)
Y  = df[60]
```

```python
X.shape
```

```python
Y.shape
```

```python
print(X)
```

# Splitting the data into test and train data


```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify = Y, random_state = 42)
```

```python
X_train.shape
```

```python
Y_train.shape
```

```python
X_test.shape
```

```python
print(X_train)
print(Y_train)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

Model Evaluation

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print('Accuracy on training data : ', training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Accuracy on test data : ', test_data_accuracy)
```

```python
import numpy as np

def predict_rock_or_mine(model, input_data):
    """
    Predict whether the given input data represents a rock or a mine.

    Parameters:
    model: The trained machine learning model
    input_data: A tuple or list of feature values

    Returns:
    str: A string indicating whether the object is a rock or a mine
    """
    # Convert the input data to a numpy array
    input_data = np.asarray(input_data)

    # Reshape the array to match the model's expected input shape
    input_data = input_data.reshape(1, -1)

    # Make the prediction
    predicted_value = model.predict(input_data)

    # Return the result
    if predicted_value[0] == 'R':
        return 'The object is a Rock'
    else:
        return 'The object is a Mine'

# Example usage
input_data = (0.0331, 0.0423, 0.0474, 0.0818, 0.0835, 0.0756, 0.0374, 0.0961, 0.0548, 0.0193, 0.0897, 0.1734,
              0.1936, 0.2803, 0.3313, 0.5020, 0.6360, 0.7096, 0.8333, 0.8730, 0.8073, 0.7507, 0.7526, 0.7298,
              0.6177, 0.4946, 0.4531, 0.4099, 0.4540, 0.4124, 0.3139, 0.3194, 0.3692, 0.3776, 0.4469, 0.4777,
              0.4716, 0.4664, 0.3893, 0.4255, 0.4064, 0.3712, 0.3863, 0.2802, 0.1283, 0.1117, 0.1303, 0.0787,
              0.0436, 0.0224, 0.0133, 0.0078, 0.0174, 0.0176, 0.0038, 0.0129, 0.0066, 0.0044, 0.0134, 0.0092)

# Assuming 'model' is your trained machine learning model
result = predict_rock_or_mine(model, input_data)
print(result)
```
